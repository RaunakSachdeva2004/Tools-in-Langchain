```python
from langchain.agents import create_agent
from langchain_classic import hub
from langgraph.prebuilt import chat_agent_executor
```

```python
import langchain
import langchain_core
import langchain_community

print(f"LangChain: {langchain.__version__}")
print(f"Core: {langchain_core.__version__}")
print(f"Community: {langchain_community.__version__}")
```

```python
# step 2 - Pull the ReAct prompt from LangChain Hub
prompt = hub.pull("hwchase17/react")  # pulls the standard ReAct agent prompt
```

```python
# Step 3: Create the ReAct agent manually with the pulled prompt
agent = create_agent(
    model=llm,
    tools=[search_tool],
    system_prompt=str(prompt)
)
```

```python
# The correct import for v1.2.0+ legacy support
from langchain_classic.agents import AgentExecutor

# Now wrap your agent
agent_executor = AgentExecutor( # agent ki baat maan krke uski baat man kr 
    agent=agent,                # ke kisi bhi step ko execute krega
    tools=[search_tool],
    verbose=True,
    handle_parsing_errors=True
)
```

```python
# Step 5: Invoke
response = agent_executor.invoke({"input": "Find the capital of Madhya Pradesh"})
print(response['output'])
```
